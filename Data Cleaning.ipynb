## Data Exploration using Seaborn

##### Import the iris data set 

```python
from sklearn.datasets import load_iris
import pandas as pd
import numpy as np
```

```python
iris = load_iris()
```

```python
iris
```

```python
data = pd.DataFrame(data=iris.data,columns=iris.feature_names)
```

```python
data.head()
```

```python
lable = pd.DataFrame(data=list(map(lambda x : iris.target_names[x],iris.target)),columns=['Species'])
```

```python
lable.head()
```

```python
iris=pd.concat([data,lable],axis=1)
```

```python
iris.head()
```

### 1. Use the displot() to see the distribution of the SepalLengthCm,SepalWidthCm,PetalLengthCm,PetalWidthCm features.Plot them as single image.

```python
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
sns.set(style='white',palette='muted',color_codes=True)
fig,axes=plt.subplots(2,2,figsize=(10,10))
sns.distplot(iris['sepal length (cm)'],hist=True,kde=True,color='b',ax=axes[0,0])
sns.distplot(iris['sepal width (cm)'],hist=True,kde=True,color='r',ax=axes[0,1])
sns.distplot(iris['petal length (cm)'],hist=True,kde=True,color='y',ax=axes[1,0])
sns.distplot(iris['petal width (cm)'],hist=True,kde=True,color='g',ax=axes[1,1])
plt.setp(axes, yticks=[])
plt.tight_layout()
plt.show()
```

### 2.DO box plot of all the features without 'Species'

```python
plt.figure(figsize=(10,10))
sns.boxplot(data=[iris['sepal length (cm)'],iris['sepal width (cm)'],iris['petal length (cm)'],iris['petal width (cm)']])
plt.title("Box Plot")
plt.xlabel("Sepal length (cm), Sepal width (cm), Petal length (cm), Petal width (cm)")
```

### 3. Do a countplot for the feature 'Species'

```python
iris.columns
```

```python
plt.figure(figsize=(10,10))
sns.countplot(x='Species',data=iris,hue='Species')
```

### 4. Do a pairplot on the features 'SepalLengthCm','SepalWidthCm','PetalLengthCm','PetalWidthCm','Species'

```python
sns.pairplot(data=iris , hue='Species',  height=2.5,)
```

### 5. Do an implot on the following SepalLengthCm,PetalLengthCm.Using hue, display the different species in different colors.

```python
plt.figure(figsize=(10,10))
sns.lmplot(data=iris,x='sepal length (cm)',y='petal length (cm)',hue='Species')
```

### 6. Do a Bar plot of 'Species' vs 'SepalLegthCm'

```python
plt.figure(figsize=(7,7))
sns.barplot(data=iris,x='Species',y='sepal length (cm)',hue='Species')
```

### 7. Using heatmap, plot the corration matrix

```python
plt.figure(figsize=(10,10))
sns.heatmap(data=iris.corr())
plt.title("co-relation")
```
